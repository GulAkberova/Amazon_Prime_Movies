```python
#Removing 'country' and 'date_added' columns
df.drop(columns=['country','date_added'],inplace=True)
df.head(2)
```

```python
#Filling missing values in 'director' and 'cast' columns with 'Unknown'
df['director'].fillna('Unknown',inplace=True)
df['cast'].fillna('Unknown',inplace=True)
df.isnull().sum()
```

```python
# Filling missing values in 'rating' column with the mode (most frequent value)
rating_values=df['rating'].value_counts()
rating_mode=df['rating'].mode()[0]
df['rating'].fillna(rating_mode,inplace=True)
df.isnull().sum()
```

```python
# Plotting the distribution of 'type' (Movie vs TV Show)
plt.figure(figsize=(8,5))
sns.countplot(data=df, x= "type", palette="Set2")
plt.title('Distribution of Content Types (Movie vs TV Show)', fontsize=14)
plt.xlabel('Type', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.show()
```

```python
#Plotting the top 10 categories in "listed_in"
top_10_categories=df['listed_in'].value_counts().head(10)
plt.figure(figsize=(10,6))
sns.barplot(x=top_10_categories.values,y=top_10_categories.index,palette="Set3")
plt.xlabel('Number of Titles', fontsize=12)
plt.ylabel('Category', fontsize=12)
plt.show()
```

```python
# Plotting the distribution of content over the release years
plt.figure(figsize=(10, 6))
sns.histplot(df['release_year'], bins=30, kde=False, color='blue')
plt.title('Distribution of Content Over the Years', fontsize=14)
plt.xlabel('Release Year', fontsize=12)
plt.ylabel('Number of Titles', fontsize=12)
plt.grid(True)
plt.show()
```

```python
# Grouping the data by 'release_year' and 'type' to observe trends over time
df_grouped = df.groupby(['release_year', 'type']).size().unstack()

# Plotting the trends over time for Movies and TV Shows
plt.figure(figsize=(10, 6))
df_grouped.plot(kind='line', marker='o', figsize=(10,6))
plt.title('Trends of Movies and TV Shows Over the Years', fontsize=14)
plt.xlabel('Release Year', fontsize=12)
plt.ylabel('Number of Titles', fontsize=12)
plt.grid(True)
plt.show()
```
